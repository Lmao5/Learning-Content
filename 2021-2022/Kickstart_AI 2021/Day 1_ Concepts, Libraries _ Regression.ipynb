```python
df = pd.read_csv('Houses.csv')
```

```python
df.head()
```

```python
feature = ['House_Age', 'Rooms', 'Bedrooms','Population', 'Households', 'Income']
feature_arr = df[feature].values
```

```python
feature_arr
```

```python
feature_arr.shape
```

```python
label_arr = df["House_Price"].values
```

```python
label_arr
```

```python
label_arr.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(feature_arr, label_arr)
```

```python
house_age = 21
rooms = 997
bedrooms = 180
population = 320
households = 98
income = 4.563
 
pred_features = [[house_age, rooms, bedrooms, population, households, income]]
```

```python
x = model.predict(pred_features)[0]

print(f'Your house is worth ${int(x//1000)}K')
```
