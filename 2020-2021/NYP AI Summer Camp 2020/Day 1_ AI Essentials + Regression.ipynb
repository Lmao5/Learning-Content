```
x = np.array([ [1], [2], [3], [4]])
y = np.array([3,5,7,9])
```

```
from sklearn.linear_model import LinearRegression
```

```
# Initialize the model to use the LinearRegression algorithm
model = LinearRegression()
```

```
# Train the model
model.fit(x,y)
```

```
# Get the learnt weights
model.coef_
```

```
# Get the learnt bias
model.intercept_
```

```
#Make predictions. Since the x array is 2d and of shape (num_egs, 1) so when i do prediction (with 2 egs) I should pass in a 2d array of shape (2,1)
my = np.array([[20], [30]])
model.predict(my)
```
