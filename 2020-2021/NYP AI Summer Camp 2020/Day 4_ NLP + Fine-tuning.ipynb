```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
#Corpus
texts =  ['It is a sunny today', 'It is rainy today', 'I hope it is not raining now']
```

```
tokenizer = Tokenizer()
```

```
#Learn the vocabulary from the corpus
#This vocabulary will be used for vectorization (Convert text to integers)
tokenizer.fit_on_texts(texts)
```

```
print(tokenizer.word_index)
```

```
new = 'It is neither sunny nor rainy today'
```

```
#Vectorize the new sentence
tokenizer.texts_to_sequences([new])
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
texts =  ['It is a sunny today', 'It is rainy today', 'I hope it is not raining now']
```

```
vect = CountVectorizer()
```

```
# Use fit instead of fit_on_texts
vect.fit(texts)
```

```
# Use vocabulary_ instead of word_index
vect.vocabulary_
```

```
new = ['It is neither sunny nor rainy today but it was sunny yesterday']
```

```
# Use transform instead of texts_to_sequences
vect.transform(new).toarray()
```

```
import pandas as pd
```

```
df = pd.read_csv('train.csv')
```

```
df.head()
```

```
#Index out the sentences
x = df['text'].values
#Index out the labels: 1 for disaster and 0 for no disaster
y = df['target'].values
```

```
print(x)
```

```
print(y[:20])
```

```
#Initialize the count vectorizer
vect = CountVectorizer()
#Fit the count vectorizer on all the sentences so you can learn a vocabulary
vect.fit(x)
```

```
#Sparse matrix is faster
x_vect = vect.transform(x)
```

```
type(x_vect)
```

```
x_np = vect.transform(x).toarray()
```

```
x_np[0]
```

```
from sklearn.linear_model import LogisticRegression
```

```
model = LogisticRegression()

model.fit(x_vect, y)
```

```
my = 'There is a forest fire'

#Perform the same preprocessing (vectorization) on my
my_vect = vect.transform([my])
model.predict(my_vect)
```

```
model_1 = LogisticRegression(C=1)
model_10 = LogisticRegression(C=10)
```

```
model_1.fit(x_vect, y)
model_1.score(x_vect, y)
```

```
model_10.fit(x_vect, y)
model_10.score(x_vect, y)
```

```
model_2 = LogisticRegression(C=5, solver='liblinear', max_iter=500)
```

```
model_s1 = LogisticRegression(solver='newton-cg')
model_s1.fit(x_vect, y)
model_s1.score(x_vect, y)
```

```
model_s2 = LogisticRegression(solver='lbfgs')
model_s2.fit(x_vect, y)
model_s2.score(x_vect, y)
```

```
model_s3 = LogisticRegression(solver='liblinear')
model_s3.fit(x_vect, y)
model_s3.score(x_vect, y)
```

```
model_s4 = LogisticRegression(solver='sag')
model_s4.fit(x_vect, y)
model_s4.score(x_vect, y)
```

```
model_s5 = LogisticRegression(solver='saga')
model_s5.fit(x_vect, y)
model_s5.score(x_vect, y)
```

```
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```
params = {
    'C':[10,50,100],
    'solver':['liblinear', 'sag'],
    'max_iter':[100,250,500]
}
```

```
model_search = LogisticRegression()
```

```
grid = GridSearchCV(model_search, params)
```

```
#RandomSearchCV
```

```
grid.fit(x_vect, y)
```

```
grid.best_params_
```

```
model_best = grid.best_estimator_
```

```
print(model_best)
```

```
#My custom sentence
my = ['Hurricane hits the road']
#Vectorizing my sentence so it becomes a set of numbers (Which can be used for making predictions)
my_vect = vect.transform(my)

#No tuning
print(model.predict(my_vect))
#Have tuning
print(model_best.predict(my_vect))
```
