```
import os
```

```
imgs = []
labels = []

for file in os.listdir('/content/dataset2'):
  #For the x array
  path = '/content/dataset2/' + file
  img = Image.open(path).convert('L')
  img = img.resize((150,150))
  imgArr = np.array(img)
  imgs.append(imgArr)


  #For the y array
  if file.startswith('cloudy'):
    labels.append(0)
  elif file.startswith('rain'):
    labels.append(1)
  elif file.startswith('shine'):
    labels.append(2)
  else:
    labels.append(3)
```

```
imgs_np = np.array(imgs)
labels_np = np.array(labels)
```
